```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import extrema
from skimage.feature import peak_local_max

xax = np.linspace(-np.pi/2, np.pi/2, 1000)
yax = np.array([(np.cos(i)*np.cos(10*i))**2 for i in xax])
image = np.array([yax for i in range(len(xax))])
dyn_range = np.max(image) - np.min(image)

h = 0.3519*dyn_range
h_maxima = extrema.h_maxima(image, h)
coordinates = peak_local_max(h_maxima, min_distance=0)

fig, ax = plt.subplots(1, 2)
ax[0].plot(coordinates[:,1], coordinates[:,0], 'r.', ms=1)
ax[0].imshow(image)
ax[0].set_title(f"h={h}")

ax[1].plot(xax, yax)
xpeaks = [i/len(xax)*np.pi-np.pi/2 for i in np.unique(coordinates[:,1])]
ypeaks = [yax[i] for i in np.unique(coordinates[:,1])]
[ax[1].axvline(i, c='r', ls='--', lw=0.5) for i in xpeaks]
[ax[1].text(i+.01, 0., f"x={i:.3f}", color='r', rotation=90) for i in xpeaks]
[ax[1].axhline(i, c='g', ls='-.', lw=0.5) for i in ypeaks]
[ax[1].text(-1, i+.01, f"y={i:.3f}", color='g') for i in ypeaks]

plt.tight_layout()

# print(f"Peaks at y={}")
```
