```python
import pandas as pd
```

```python
fin = "../rank_neighbours/output/ranked_1000_neigh.dat"
data = pd.read_csv(fin, sep='\t', skiprows=1, names=["idx", "radius", "x", "y", "z"])
```

```python
import matplotlib.pyplot as plt

n = 100
plt.loglog([x for x in range(len(data))][::n], data['radius'][::n])
plt.xlabel("particle no")
plt.ylabel("radius for 1000 neigh")
```

```python
import numpy as np

n = 100
plt.loglog([x for x in range(len(data))][::n], np.gradient(data['radius'])[::n])
plt.xlabel("particle no")
plt.ylabel("gradient")
```

```python
n = 1
hist = np.histogram(data['radius'][::n])
plt.hist(data['radius'][::n], bins=np.logspace(np.log10(min(data['radius'][::n])),np.log10(max(data['radius'][::n])),100), histtype='bar')
plt.xlabel("radius")
plt.ylabel("count")
plt.xscale('log')
```
